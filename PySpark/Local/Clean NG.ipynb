```python
from pathlib import Path
import os
import glob2

def getindex(cont,idx):
    try:
        return cont.index(idx)
    except:
        return -1
```

```python

def cleanFile(fin, fout):
    text = Path(fin).read_text()

    lines = text.split('\n')
    
    myfile = open(fout, 'w')
    
    for line in text.split('\n'):
        words = line.split(' ')
        if words[0][-1:]!=':':
            newline = ' '.join([word for word in words if getindex(word,'.')==-1 or getindex(word,'.')==len(word)-1])
            myfile.write(newline + '\n')

    myfile.close()
    
```

```python
for folder in ['alt.atheism','comp.graphics','comp.os.ms-windows.misc','comp.sys.ibm.pc.hardware','comp.sys.mac.hardware','comp.windows.x','misc.forsale','rec.autos','rec.motorcycles','rec.sport.baseball','rec.sport.hockey','sci.crypt','sci.electronics','sci.med','sci.space','soc.religion.christian','talk.politics.guns','talk.politics.mideast','talk.politics.misc','talk.religion.misc']:
    p = 'C:\\20_newsgroups\\files\\' + folder + '\\*'
    o = 'C:\\20_newsgroups\\clean\\' + folder + '\\'
    f_list = glob2.glob(p)

    for fin in f_list:
        #print(fin)
        fout = o + os.path.basename(fin)
        #print(fout)
        cleanFile(fin,fout)
```
